# CSV path 가져오기

```python
def find_all_csv_files(data_dir):
    file_paths = []
    for root, dirs, files in os.walk(data_dir):
        csv_files = [f for f in files if f.endswith('.csv')]
        for csv_file in csv_files:
            file_path = os.path.join(root, csv_file)
            file_paths.append(file_path)
    return file_paths
```

```python
csv_files = find_all_csv_files(data_dir)
print(len(csv_files))
print(csv_files)
```

```python
hosp_csv_files = find_all_csv_files(hosp_dir_path)
print(len(hosp_csv_files))
print(hosp_csv_files)
```

```python
icu_csv_files = find_all_csv_files(icu_dir_path)
print(len(icu_csv_files))
print(icu_csv_files)
```

# 기본적인 EDA


```python
from IPython.display import display

def basic_eda(file_paths):
    for file_path in file_paths:
        try:
            # 파일 경로 출력
            print(f"\nPerforming EDA on {file_path}...\n")
            
            # CSV 파일 읽기
            try:
                # 처음 시도: 일반적으로 파일 읽기
                df = pd.read_csv(file_path)
            except MemoryError:
                chunk_size = 10000000  # 1천만 읽기
                print(f"Memory error encountered. Reloading with chunk size: {chunk_size}")
                chunk_reader = pd.read_csv(file_path, chunksize=chunk_size, low_memory=True)
                df = next(chunk_reader)
            
            # 데이터 크기 (행, 열) 출력
            print(f"Shape of the DataFrame: {df.shape}\n")
            
            # 첫 5개 행 출력 (Jupyter Notebook 스타일로)
            print("First 5 rows of the DataFrame:")
            display(df.head())  # 이쁘게 출력
            
            # 열별 결측값(null 값) 개수 출력
            print("Missing values in each column:")
            display(df.isnull().sum())  # 이쁘게 출력
            
            # 데이터 타입 정보 출력
            print("Data types of each column:")
            display(df.dtypes)  # 이쁘게 출력
            
            # 기술 통계 요약 (수치형 데이터)
            print("Summary statistics (numerical columns):")
            display(df.describe())  # 이쁘게 출력
            
        except Exception as e:
            print(f"An error occurred while processing {file_path}: {e}")
```

## hosp
